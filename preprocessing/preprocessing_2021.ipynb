```python
import pandas as pd
import numpy as np
```

```python
df2021 = pd.read_csv('Odata2021File.csv', delimiter=';')
df2021
```

```python
cols = df2021.columns.values.tolist()
for k, v in enumerate(cols):
    print(f'{k} : {v}')
```

```python
to_drop = [cols[0]] + cols[3:9] + [cols[10]] + [cols[15]]
df2021.drop(to_drop, axis=1, inplace=True)
df2021
```

```python
df2021 = df2021[df2021['ClassLangName']=='українська']
print(df2021['ClassLangName'].unique())
df2021.drop(columns=['ClassLangName'],axis=1, inplace=True)
df2021
```

```python
df2021 = df2021[df2021['MathLang']=='українська']
print(df2021['MathLang'].unique())
```

```python
df2021 = df2021[~df2021['UMLTestStatus'].isin(["Не з’явився", "Анульовано"])]
df2021
```

```python
df2021.drop(columns=['UMLTestStatus', 'UkrTestStatus', 'UkrBall100', 'HistLang', 'MathLang', 'MathTestStatus', 'MathStLang', 'MathStTestStatus', 'PhysLang', 'ChemLang', 'BioLang', 'GeoLang', ], axis=1, inplace=True)
df2021
```

```python
df2021.drop(columns=['GeoBall100'], axis=1, inplace=True)
df2021.rename(columns={'SexTypeName':'SexType', 'EOTypeName':'SchoolType', 'EORegName':'SchoolRegion', 'EOAreaName':'SchoolArea', 'UMLBall100':'Ukr', 'HistBall100':'Hist', 'MathBall100':'Math', 'PhysBall100':'Phys', 'ChemBall100':'Chem', 'BioBall100':'Bio', 'EngBall100':'Eng', 'FraBall100':'Fra', 'DeuBall100':'Deu', 'SpaBall100':'Spa'}, inplace=True)
df2021['SexType'] = df2021['SexType'].apply(lambda x: 1 if x == 'чоловіча' else 0)
df2021['TestType'] = 'ZNO2021'
df2021
```

```python
urban, village, nan = 0, 0, 0
values = list(df2021['EOTerName'].values)
print(len(values))
for occurrence in df2021['EOTerName']:
       # occurrence = str(occurrence)
       if 'nan' in str(occurrence) or 'NaN' in str(occurrence):
              # values.remove(occurrence)
              nan += 1
       if 'м.' in str(occurrence) or 'район міста' in str(occurrence):
              # values.remove(occurrence)
              urban += 1
       if 'с.' in str(occurrence) or 'смт ' in str(occurrence) or 'смт.' in str(occurrence) or 'с-ще ' in str(occurrence):
              # values.remove(occurrence)
              village += 1
print(nan)
print(urban)
print(village)
print(nan + urban + village)
```

```python
df2021['Urban'] = np.NaN
urban_mask = df2021['EOTerName'].str.contains('м\.|район міста', case=False, na=False)
village_mask = df2021['EOTerName'].apply(lambda x: 'с.' in str(x) or 'смт ' in str(x) or 'смт.' in str(x) or 'с-ще ' in str(x))
df2021.loc[urban_mask, 'Urban'] = 1
df2021.loc[village_mask, 'Urban'] = 0
df2021
```

```python
df2021.drop(columns=['EOTerName'], inplace=True)
df2021.to_csv('data2021.csv', index=False, sep=';', encoding='utf-8')
df2021.to_parquet('data2021_parquet.parquet', index=False)
```
