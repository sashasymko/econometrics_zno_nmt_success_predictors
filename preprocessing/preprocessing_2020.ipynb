```python
import pandas as pd
import numpy as np
```

```python
df2020 = pd.read_csv('Odata2020File.csv', delimiter=';')
df2020
```

```python
df2020.drop(columns='REGNAME', axis=1, inplace=True)
df2020 = df2020[df2020['ClassLangName']=='українська']
print(df2020['ClassLangName'].unique())
df2020.drop(columns=['ClassLangName'],axis=1, inplace=True)
df2020
```

```python
df2020 = df2020[df2020['mathLang'].isin(['українська', pd.NA])]
print(df2020['mathLang'].unique())
print(df2020['HistLang'].unique())
```

```python
df2020 = df2020[df2020['HistLang'].isin(['українська', pd.NA])]
print(df2020['HistLang'].unique())
df2020
```

```python
df2020 = df2020[~df2020['UkrTestStatus'].isin(["Не з’явився", "Анульовано"])]
print(df2020['UkrTestStatus'].unique())
df2020
```

```python
df2020.drop(columns=['UkrTestStatus', 'HistLang', 'mathLang', 'geoBall100'], axis=1, inplace=True)
df2020
```

```python
df2020.rename(columns={'SEXTYPENAME':'SexType', 'EOTYPENAME':'SchoolType', 'EORegName':'SchoolRegion', 'EOAreaName':'SchoolArea', 'UkrBall100':'Ukr', 'histBall100':'Hist', 'mathBall100':'Math', 'physBall100':'Phys', 'chemBall100':'Chem', 'bioBall100':'Bio', 'engBall100':'Eng', 'fraBall100':'Fra', 'deuBall100':'Deu', 'spaBall100':'Spa'}, inplace=True)
df2020['SexType'] = df2020['SexType'].apply(lambda x: 1 if x == 'чоловіча' else 0)
df2020['TestType'] = 'ZNO2020'
df2020.drop(columns=['EONAME'], axis=1, inplace=True)
df2020
```

```python
urban, village, nan = 0, 0, 0
values = list(df2020['EOTerName'].values)
print(len(values))
for occurrence in df2020['EOTerName']:
       # occurrence = str(occurrence)
       if 'nan' in str(occurrence) or 'NaN' in str(occurrence):
              # values.remove(occurrence)
              nan += 1
       if 'м.' in str(occurrence) or 'район міста' in str(occurrence):
              # values.remove(occurrence)
              urban += 1
       if 'с.' in str(occurrence) or 'смт ' in str(occurrence) or 'смт.' in str(occurrence) or 'с-ще ' in str(occurrence):
              # values.remove(occurrence)
              village += 1
print(nan)
print(urban)
print(village)
print(nan + urban + village)
```

```python
df2020['Urban'] = np.NaN
urban_mask = df2020['EOTerName'].str.contains('м\.|район міста', case=False, na=False)
village_mask = df2020['EOTerName'].apply(lambda x: 'с.' in str(x) or 'смт ' in str(x) or 'смт.' in str(x) or 'с-ще ' in str(x))
df2020.loc[urban_mask, 'Urban'] = 1
df2020.loc[village_mask, 'Urban'] = 0
df2020
```

```python
df2020.drop(columns=['EOTerName'], inplace=True)
df2020.to_csv('data2020.csv', index=False, sep=';', encoding='utf-8')
df2020.to_parquet('data2020_parquet.parquet', index=False)
```
