```python
import pandas as pd
import numpy as np
```

```python
df2022 = pd.read_csv('Odata2022File.csv', delimiter=';')
df2022
```

```python
columns_to_drop = ['OUTID', 'RegName', 'AREANAME', 'TERNAME', 'RegTypeName', 'TerTypeName', 'EONAME', 'EOParent', 'TestDate', 'Block1', 'Block1Ball', 'Block2', 'Block2Ball', 'Block3Ball', 'Block3', 'PTRegName', 'PTAreaName', 'PTTerName']
df2022.drop(columns_to_drop, axis=1, inplace=True)
df2022
```

```python
df2022.rename(columns={'SexTypeName':'SexType', 'EOTypeName': 'SchoolType', 'EORegName':'SchoolRegion', 'EOAreaName':'SchoolArea', 'Test':'TestType', 'Block1':'Block1', 'Block1Ball100':'Ukr', 'Block2Ball100':'Hist', 'Block3Ball100':'Math', 'TestStatus':'Status'}, inplace=True)
df2022
```

```python
df2022 = df2022[~df2022['Status'].isin(["Не з’явився", "Анульовано"])]
df2022['SexType'] = df2022['SexType'].apply(lambda x: 1 if x == 'чоловіча' else 0)
df2022['TestType'] = 'NMT2022'
df2022['Status'] = df2022['Status'].apply(lambda x: 1 if x == 'Зараховано' else 0)
df2022
```

```python
urban, village, nan = 0, 0, 0
values = list(df2022['EOTerName'].values)
print(len(values))
for occurrence in df2022['EOTerName']:
       # occurrence = str(occurrence)
       if 'nan' in str(occurrence) or 'NaN' in str(occurrence):
              # values.remove(occurrence)
              nan += 1
       if 'м.' in str(occurrence) or 'район міста' in str(occurrence):
              # values.remove(occurrence)
              urban += 1
       if 'с.' in str(occurrence) or 'смт ' in str(occurrence) or 'смт.' in str(occurrence) or 'с-ще ' in str(occurrence):
              # values.remove(occurrence)
              village += 1
print(nan)
print(urban)
print(village)
print(nan + urban + village)
```

```python
df2022['Phys'] = np.NaN
df2022['Chem'] = np.NaN
df2022['Bio'] = np.NaN
df2022['Eng'] = np.NaN
df2022['Fra'] = np.NaN
df2022['Deu'] = np.NaN
df2022['Spa'] = np.NaN
df2022
```

```python
df2022['Urban'] = np.NaN
urban_mask = df2022['EOTerName'].str.contains('м\.|район міста', case=False, na=False)
village_mask = df2022['EOTerName'].apply(lambda x: 'с.' in str(x) or 'смт ' in str(x) or 'смт.' in str(x) or 'с-ще ' in str(x))
df2022.loc[urban_mask, 'Urban'] = 1
df2022.loc[village_mask, 'Urban'] = 0
df2022
```

```python
df2022.drop(columns='EOTerName', inplace=True)
df2022
```

```python
df2022.drop(columns='Status', axis=1, inplace=True)
df2022.to_csv('data2022.csv', index=False, sep=';', encoding='utf-8')
df2022.to_parquet('data2022_parquet.parquet', index=False)
```
