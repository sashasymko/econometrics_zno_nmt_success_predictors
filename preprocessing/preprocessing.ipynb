Зробити:
- видалити outid, RegName, AreaName, TerName, RegTypeName, TerTypeName, EOName, EOParent, TestDate, UkrBlock (AnyBlock), UkrBlockBall (AnyBlockBall), PTRegName, PRAreaName, PTTerName
- SexTypeName = SexType (1 - male), EOTerName = Urban (1 - true), Test = TestType(NMT2023), UrkBlockBall100 = Ukr (AnyBlockBall100 = Any), 
- якщо UkrBlockStatus (чи anyStatus) = 'Не зʼявився' або "Анульовано", то дропнути рядок 


```python
print(df2023.columns)
columns_to_drop = ['outid', 'RegName', 'AreaName', 'TerName', 'RegTypeName',
       'TerTypeName', 'EOName', 'EOTerName', 'EOParent', 'PTRegName',
       'PTAreaName', 'PTTerName']
df2023.drop(columns_to_drop, axis=1, inplace=True)
df2023
```

```python
df2023.drop('TestDate', axis=1, inplace=True)
df2023['Test'] = 'NMT2023'
df2023.rename(columns={'Test': 'TestType'}, inplace=True)
columns_to_drop = []
columns = list(df2023.columns)[6:]
for i in range(len(columns)):
       if (i % 4 == 0 or i % 4 == 3):
              columns_to_drop.append(columns[i])
print(columns_to_drop)
df2023
```

```python
df2023.drop(columns_to_drop, axis=1, inplace=True)
df2023
```

```python
df2023['SexTypeName'] = df2023['SexTypeName'].apply(lambda x: 1 if x == 'чоловіча' else 0)
df2023.rename(columns={'SexTypeName': 'SexType'}, inplace=True)
columns = list(df2023.columns[8:])
columns_to_drop = []
for i in range(len(columns)):
       if (i % 2 == 0):
              columns_to_drop.append(columns[i])
print(columns_to_drop)
df2023
```

```python
df2023.drop(columns_to_drop, axis=1, inplace=True)
df2023
```

```python
rename_part = 'BlockBall100'
meaningful_part = ['Ukr', 'Hist', 'Math', 'Phys', 'Chem', 'Bio', 'Eng', 'Fra', 'Deu', 'Spa']
old_names = [name+rename_part for name in meaningful_part]
to_rename = dict(zip(old_names, meaningful_part))
df2023.rename(columns=to_rename, inplace=True)
df2023
```

```python
df2023.rename(columns={'UkrBlockStatus':'Status'}, inplace=True)
df2023 = df2023[~df2023['Status'].isin(["Не з’явився", "Анульовано"])]
print(df2023['Status'].unique())
df2023
```

```python
df2023['Status'] = df2023['Status'].apply(lambda x: 1 if x == 'Зараховано' else 0)
df2023
```

```python
df2023.rename(columns={'EORegName': 'SchoolRegion', 'EOAreaName':'SchoolArea', 'EOTypeName':'SchoolType'}, inplace=True)
df2023
```

```python
df2023_original = pd.read_csv('Odata2023File.csv', delimiter=';')
df2023_original
```

```python
urban, village, nan = 0, 0, 0
df2023_original = df2023_original[~df2023_original['UkrBlockStatus'].isin(["Не з’явився", "Анульовано"])]
values = list(df2023_original['EOTerName'].values)
for occurrence in df2023_original['EOTerName']:
       # occurrence = str(occurrence)
       if 'nan' in str(occurrence) or 'NaN' in str(occurrence):
              # values.remove(occurrence)
              nan += 1
       if 'м.' in str(occurrence) or 'район міста' in str(occurrence):
              # values.remove(occurrence)
              urban += 1
       if 'с.' in str(occurrence) or 'смт ' in str(occurrence) or 'смт.' in str(occurrence) or 'с-ще ' in str(occurrence):
              # values.remove(occurrence)
              village += 1
print(nan)
print(urban)
print(village)
```

```python
df2023['Urban'] = np.NaN
df2023
```

```python
urban_mask = df2023_original['EOTerName'].str.contains('м\.|район міста', case=False, na=False)
village_mask = df2023_original['EOTerName'].apply(lambda x: 'с.' in str(x) or 'смт ' in str(x) or 'смт.' in str(x) or 'с-ще ' in str(x))
df2023.loc[urban_mask, 'Urban'] = 1
df2023.loc[village_mask, 'Urban'] = 0
df2023
```
